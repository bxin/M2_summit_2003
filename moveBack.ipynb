```python
from astropy.time import Time
from datetime import timedelta
from lsst_efd_client import EfdClient
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
client = EfdClient('summit_efd')
csc_index = 1
```

```python
#start = Time('2020-03-04T19:25:00')
#start = Time('2020-03-04T23:42:00') # -x motion limit
#start = Time('2020-03-05T00:07:00') # +y motion limit
#start = Time('2020-03-05T00:22:00') # -y motion limit
#start = Time('2020-03-05T00:36:00') # +z motion limit
start = Time('2020-03-05T00:49:00') # -z motion limit
end = start+ timedelta(minutes = 5)
```

```python
end.tai
```

```python
dfa = await client.select_time_series('lsst.sal.MTM2.tangentActuatorAbsolutePositionSteps', '*', start.tai, end.tai, csc_index)
dfa.head()
```

```python
dfa.tail()
```

```python
df = await client.select_time_series('lsst.sal.MTM2.axialActuatorAbsolutePositionSteps', '*', start.tai, end.tai, csc_index)
df.head()
```

```python
df.tail()
```

```python
nloop = 1000
aa = np.zeros((300, 78)) #we've done a few tests, we divided the recovery into 1000 iterations. We only needed 300
```

```python
dfa['tangentLink0DegAbsolutePositionSteps'][-1]-dfa['tangentLink0DegAbsolutePositionSteps'][0]
```

```python
for i in range(72):
    s = df['axialAbsolutePositionSteps%d'%i][-1] - df['axialAbsolutePositionSteps%d'%i][0]
    aa[:,i] = -s/nloop
for i in range(72,78):
    s = dfa['tangentLink%dDegAbsolutePositionSteps'%((i-72)*60)][-1]-dfa['tangentLink%dDegAbsolutePositionSteps'%((i-72)*60)][0]
    aa[:,i] = -s/nloop
```

```python
aa[:2,:]
```

```python
#fid = open('../m2csv/moveBack300.csv', 'w') #-x motion limit
#fid = open('../m2csv/moveBack_yp.csv', 'w') #+y motion limit
#fid = open('../m2csv/moveBack_ym.csv', 'w') #-y motion limit
#fid = open('../m2csv/moveBack_zp.csv', 'w') #+z motion limit
fid = open('../m2csv/moveBack_zm.csv', 'w') #+z motion limit

for i in range(aa.shape[0]):
    fid.write('S ') #these are steps. Use "F" for forces
    for j in range(aa.shape[1]):
        fid.write('%.10f'%aa[i,j])
        if j<aa.shape[1]-1:
            fid.write(',')
        else:
            fid.write('\n')
fid.close()
```
